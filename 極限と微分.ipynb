### 1_1接線の描写
- 導関数を手打ちしてみる

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def my_func(x):
    return 3*x**2 + 4*x - 5

def my_func_dif(x):
    return 6*x + 4

x = np.linspace(-3,3)
y = my_func(x)

a = 1

y_t =my_func_dif(a)*x + my_func(a) - my_func_dif(a)*a

plt.plot(x,y,label="y")
plt.plot(x,y_t,label="y_t")
plt.legend()

plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()

plt.show()
```

### シグモイド関数
- $y=\frac{1}{1+exp(-1)}$
- 上記で定義される関数は、ネットワーク解析で使われる。
- $\frac{dy}{dx}=(1-y)y$
- 微分の形の実装を試みる。

```python
import numpy as np
import matplotlib.pylab as plt

def sigmoid_func(x):
    return 1/(1+np.exp(-x))

def grad_sigmoid(x):
    y = sigmoid_func(x)
    return (1-y)*y

x = np.linspace(-5,5)
y = sigmoid_func(x)
y_grad = grad_sigmoid(x)

plt.plot(x,y,label="y")
plt.plot(x,y_grad,label="y")
plt.legend()

plt.xlabel("x",size=14)
plt.ylabel("y",size=14)
plt.grid()

plt.show()
```

### 最急降下法の実装
- $f(x) = x^2 - 2x$
- $\frac{df(x)}{dx} = 2x - 2$

- 上記の関数の最小値を、最急降下法を使って求める。

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def my_func(x):
    return x**2 - 2*x

def grad_func(x):
    return 2*x - 2

eta = 0.1 #学習関数
x = 4.0 #xに初期値を設定
record_x = []#xの記録
record_y = []#yの記録

for i in range(20):
    y = my_func(x)
    record_x.append(x)
    record_y.append(y)
    x -= eta * grad_func(x)
    
x_f = np.linspace(-2,4)
y_f = my_func(x_f)

plt.plot(x_f,y_f,linestyle="dashed") #関数を点線で表示
```
