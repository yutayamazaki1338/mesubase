```python
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests

#基礎データ構築
#100件表示されるベースページの情報をまとめる
def basepage_basedatamake():
    
    res = requests.get(base_url)
    soup = BeautifulSoup(res.text)

    list = soup.select("div.infodatabox-box-txt")
    d=0
    A=[]
    B=[]
    for a in list:
        if d !=6:
            A.append(a.get_text().split())
            d += 1
        elif d == 6:
            B.append(A)
            A = []
            A.append(a.get_text().split())
            d = 1
    B.append(A) #調整用

    for i in range(len(B)):
        zyuusyo.append(B[i][0][0])
        moyori.append(B[i][1][0])
        kyori.append(B[i][2][0])
        tinryou.append(B[i][3][1])
        kanri.append(B[i][3][3])
        sikikin.append(B[i][3][5])
        menseki.append(B[i][4][1])
        madori.append(B[i][4][3])
        tiku.append(B[i][5][0])

        
#次のページのURLを探してくる        
def search_URL():
    res = requests.get(base_url)
    soup = BeautifulSoup(res.text)

    A6 = soup.select("p.pagination-parts > a")
    return open_url + A6[1]["href"]#次のページのリンクに更新

#詳細情報のページのリンクを探してくる
def detail_url():
    res = requests.get(base_url)
    soup = BeautifulSoup(res.text)
    list = soup.select("h2.cassettebox-title > a")
    info_url = []
    for a in list:
        info_url.append(a["href"])
    return info_url

#詳細情報を探してくる
def detail_datamake(b):
    a=0
    
    for i in range(100):
        info1_url = open_url + info_url[i]#一つ目の物件表示

        res = requests.get(info1_url)
        soup = BeautifulSoup(res.text)
        
        #方角
        try:
            A1 = soup.select("div.property_data-body")
            hougaku.append(A1[6].get_text().split()[0])
        except IndexError:
            hougaku.append("-")
            print("エラー検知、ページが存在しません")

            #階層
        try:
            A2 = soup.select("table.data_table")
            syozaikai.append(A2[0].get_text().split()[5])
        except IndexError:
            hougaku.append("-")

            #構造
        try:
            A3 = soup.select("table.data_table")
            kouzou.append(A3[0].get_text().split()[3])
        except IndexError:
            hougaku.append("-")    

            #部屋の特徴
        try:
            A4 = soup.select("div.bgc-wht>ul")
            tokutyou.append(A4[0].get_text().split()[0])
        except IndexError:
            hougaku.append("-")

            
            #周辺情報
        try:
            H = []
            A5 = soup.select("td.data_around > ul")
            for s in range(len(A5)):
                H.append(A5[s].get_text().split())
                syuuhen.append(H[0])
        except IndexError:
            hougaku.append("-")
                
        time.sleep(1)
        a+=1
        print(a,b
             )



    
    
#初期条件    
base_url = 'https://suumo.jp/jj/common/ichiran/JJ901FC004/?initFlg=1&seniFlg=1&ar=030&ra=030013&rnTmp=0005&ct=9999999&cb=0.0&kt=9999999&xt=9999999&et=9999999&cn=9999999&newflg=0&km=1&rn=0005&pc=100'#現在のページ
open_url="https://suumo.jp"#飛ぶのに必要なリンクとくっつける

zyuusyo = []
moyori = []
kyori = []
tinryou = []
kanri = []
sikikin =[]
menseki =[]
madori =[]
tiku =[]
hougaku = []
syozaikai = []
kouzou = []
tokutyou = []
syuuhen = []

b = 1 #ベースページの更新状況

A = pd.DataFrame({"住所":zyuusyo, "最寄り駅":moyori, "最寄り駅までの距離":kyori, "賃料":tinryou,
                   "維持費・管理費":kanri, "礼金/敷金":sikikin, "敷地面積":menseki, "間取り":madori,
                   "築年数":tiku, "方角":hougaku, "所在階/建物回数":syozaikai, "構造":kouzou,
                  "部屋の特徴":tokutyou, "周辺情報":syuuhen})
```
